```python
# Imports necessary modules
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

# Loads data from CSV and approximates no. of bins using the sqrt rule
data = pd.read_csv('/content/sample_data/fuel_econ.csv')
bins_sqrt = int(np.sqrt(len(data)))

# Plots and labels histogram using the data
sns.histplot(data["co2"], bins=bins_sqrt, kde=True)
plt.xlabel('CO2')
plt.ylabel('Frequency')
plt.title('Distribution of CO2')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

data = pd. read_csv('/content/sample_data/fuel_econ.csv')
numeric_data = data.select_dtypes(include=[np.number])
corr_matrix = numeric_data.corr()

sns.heatmap(corr_matrix, annot=False, cmap='RdBu', vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.show()
```

**Data Interpretation**

***a. Histograms***

The CO2 Histogram is symmetric. This means the data has a bell curve, with most of the data clustered between around 250 to 450 and peaks at around 300. The data ranges from 0 to 700, having a wide spread.

The Highway Histogram is skewed right. This means most of the data is found on the lower side of the graph, specifically between 20 and 40 and peaking around 30. Moreover, the spread of the data is narrow with a minimum of around 10 and a maximum of 60.

The feScore Histogram is symmetric. The data is distributed normally with a majority of the data clustered between 4 and 8 and peaking at around 5. The data has a very narrow spread with values ranging from 0 to 10.


***b. Heatmap***

The data in the heatmap forms a number of clusters. Data such as City and UCity is highly correlated with Highways and UHighways. A similar positive correlation can be seen between id and Year as well as feScore and ghgScore. An interesting observation is that pv4 has little to no correlation with any data except pv2.
